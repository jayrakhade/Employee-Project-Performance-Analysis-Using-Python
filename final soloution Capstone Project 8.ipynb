```python
#Importing necessary Libraries which will require to do the entire project.
import numpy as np
import pandas as pd
```

```python
#Creating Series of lists to use them as columns for the dataframes

id = pd.Series(['A001','A002','A003','A004','A005'])

project = pd.Series(['Project 1', 'Project 2', 'Project 3', 'Project 4', 'Project 5','Project 6', 'Project 7', 'Project 8', 'Project 9', 'Project 10', 
                     'Project 11', 'Project 12', 'Project 13', 'Project 14'])

name = pd.Series(["John Alter", "Alice Luxumberg", "Tom Sabestine", "Nina Adgra", "Amy Johny"])

gender = pd.Series(["M", "F", "M", "F", "F"])

city = pd.Series(["Paris", "London", "Berlin", "New York", "Madrid"])

age = pd.Series([25,27,29,31,30])

status = ["Finished", "Ongoing", "Failed"]

designation_level = [2,2,3,2,3]

status_list = pd.Series(["Finished", "Ongoing","Finished", "Ongoing", "Finished", "Failed", "Finished", "Failed", "Ongoing", "Finished", "Failed",
                         "Ongoing", "Finished", "Finished" ])
```

```python
# Creating the Employee DataFrame...

emp_df = pd.DataFrame ({
    "ID" : id,
    "Name" : name,
    "Gender" : gender,
    "City" : city,
    "Age" : age
})

print("Employee DataFrame: ")
emp_df
```

```python
#Creating the Seniority Level DataFrame...

senio_df = pd.DataFrame({
    "ID" : id,
    "Designation Level" : designation_level
})

print("Seniority Level DataFrame: ")
senio_df
```

```python
#Creating the Project DataFrame...

proj_df = pd.DataFrame({
    "ID" : pd.Series(["A001", "A002", "A003", "A004", "A005", "A002", "A005", "A003", "A001", "A003", "A001", "A004", "A004", "A005"]),
    "Project" : project,
    "Cost" : pd.Series([1002000, 2000000, 4500000, 5500000, None, 680000, 400000, 350000, None, 300000, 2000000, 1000000, 3000000, 200000]),
    "Status" : pd.Series(["Finished", "Ongoing", "Finished", "Ongoing", "Finished", "Failed", "Finished", "Failed", "Ongoing", "Finished",
                          "Failed", "Ongoing", "Finished", "Finished"])
})

print("Project DataFrame: ")
proj_df
```

```python
#Saving the DataFrames into .CSV files...

emp_df.to_csv("employee_dataframe.csv" , index=False)
senio_df.to_csv("seniority_level_dataframe.csv", index=False)
proj_df.to_csv("project_dataframe.csv", index=False)
```

```python
#Creating a list of running averages using 'Cost' column in the project dataframe...

running = 0
count = 0

# Calculating the average cost...
average_cost = round(proj_df['Cost'].mean(), 2)

# Replacing missing values with average cost using a for loop...
for i in range(len(proj_df)):
    if pd.isnull(proj_df.loc[i, 'Cost']):       #isnull() function is used to check if there is any null values...
        proj_df.loc[i, 'Cost'] = average_cost

    #else function is used to iterate the loop if there is no missing values...
    else:
        running += proj_df.loc[i, 'Cost']
        count += 1

print("Updated Project DataFrame:")
print(proj_df)
```

```python
#Usign .split() function to split the name...
empl_df = emp_df["Name"].str.split()

print("Employee DataFrame: ")
empl_df
```

```python
#splitting the `"Name"` column into two different columns...

emp_df[["First Name", "Last Name"]] = emp_df["Name"].str.split(' ', expand=True)

#dropping the "Name" column from theoriginal DataFrame using ,drop() method...
emp_df.drop("Name", axis = 1, inplace=True)

print("Employee DataFrame: ")
emp_df
```

```python
#re-indexing `First Name` column...
cols = list(emp_df.columns)

#inserting the column in 1st index...
cols.insert(1, cols.pop(4))

#used to reorder the columns of a DataFrame.
emp_df = emp_df.reindex(columns=cols)

print("Employee DataFrame: ")
emp_df
```

```python
#re-indexing 'Last Name' column...

cols = list(emp_df.columns)

#inserting the column in 2nd index...
cols.insert(2, cols.pop(5))

#used to reorder the columns of a DataFrame...
emp_df = emp_df.reindex(columns=cols)

print("Employee DataFrame: ")
emp_df
```

```python
#using pd.merge() function, to merge the dataframes.
#We can merge two dataframes at a time.

semi_final = pd.merge(emp_df, proj_df, on="ID")

print("Semi-Final DataFrame: ")
semi_final
```

```python
#Merging of semi-final dataframe with seniority level dataframe, and naming it as Final DataFrame...
final = pd.merge(semi_final, senio_df, on="ID")

final.to_csv("Final_Dataframe.csv", index=False)

print("Final DataFrame: ")
final
```

```python
# Creating a copy of final DataFrame...
final_copy = final

print("Copy of Final DataFrame: ")
final_copy
```

```python
#Using 'for' loop to creating a list of Bonus cost
#Later, inserted the list into the final dataframe
results_list=[]

for i, r in final_copy.iterrows():
    if(final_copy['Status'].iat[i] == 'Finished'):             #checking every index if "Finished" is present at any index of status column...
        result = round((r['Cost']) + (r['Cost']) * 0.05, 2)    #if found, then bonus = cost + 5% of cost... also rounding the bonus upto 2 decimal points
        results_list.append(result)                            #append() will add the result in the end of the list... 
    else:
        result = round((r['Cost']) + (r['Cost']) * 0, 2)       #if not, then no bonus...
        results_list.append(result)

final_copy.insert(9, "Bonus-5%", results_list)                 #inserting the bonus-5% column into the final Dataframe at index 9...

print("Updated Final DataFrame: ")
final_copy
```

```python
#As per the task given, I used 'for' loop to create a list of new designations
#Later,inserted the list into the final dataframe

new_designation = []

for i, r in final_copy.iterrows():
    #checking if the status of the project is 'Failed'...
    if (final_copy["Status"].iat[i] == 'Failed'):
        result = (r['Designation Level']) - 1    #if 'Failed' then demote 1 level from designation level...
        new_designation.append(result)
    else:
        result = (r['Designation Level']) - 0    #if not 'Failed' then do not demote any level from designation level...
        new_designation.append(result)

#inesrting the new designation column...
final_copy.insert(10, 'Designation', new_designation)     #inserting the Designation column into the final Dataframe at index 10...

print("Updated Final DataFrame: ")
final_copy
```

```python
#droping the previous designation level...
final_copy.drop("Designation Level", axis=1 , inplace=True)
```

```python
#DataFrame with updated and renamed Designation Levels

final_copy.rename(columns={'Designation': 'Designation Level'}, inplace=True)

print("Updated Final DataFrame: ")
final_copy
```

```python
#deleting the employee records having designation level > 4...

to_delete = 0

for index, row in final_copy.iterrows():
    if row["Designation Level"] > 4:                      #checking if any of the employees have designation level > 4...
        to_delete.append(index)                           #if found, then adding the index at the end of the list `to_delete`...
        final_copy.drop(to_delete, inplace=True)          #then, removing the index from the Dataframe...

print("Updated Final DataFrame: ")
final_copy
```

```python
#Adding "Mr." to the first name of males and "Mrs." to the first name of females

final_copy['First Name'] = np.where(final_copy['Gender'] == 'M', 'Mr. ' + final_copy['First Name'], 'Mrs. ' + final_copy['First Name'])

#dropping the Gender column from the dataframe...
final_copy = final_copy.drop("Gender", axis=1)

print("Updated Final DataFrame: ")
final_copy
```

```python
#used iteration of rows and if function to promote designation level by 1 for the wmployees having age > 29...

for i, r in final_copy.iterrows():
    if row["Age"] > 29:
        final_copy.loc[i, 'Designation Level'] += 1       #if any employee has age > 29, then promoting designation level by 1...

    else:
        final_copy.loc[i, 'Designation Level'] += 0       #if any employee has age !> 29, then designation level remains same...

print("Updated Final DataFrame: ")
```

```python
#Grouping by employee ID and first name, and calculating the total project cost by summing up the cost column...
TotalProjCost = final_copy.groupby(['ID', 'First Name'])['Cost'].sum().reset_index()

#Renaming the columns...
TotalProjCost.columns = ['ID', 'First Name', 'Total Cost']

#converting the TotalProjCost Dataframe  to a .csv file...
TotalProjCost.to_csv("Total_Project_Cost_Dataframe.csv")

print("TotalProjCost DataFrame: ")
TotalProjCost
```
